Pipeline Results

```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
IMG_DIR = "./laptop" 

pipeline = [
    ("canny_pip.png",        "Canny Edge Pipeline"),
    ("hough.png",            "Hough Line Detection"),
    ("extended_lines.png",   "Extended Lines"),
    ("edge_points.png",      "Edge Points"),
    ("final_edge.png",       "Final Edge Map"),
    ("corner_roi.png",       "Corner ROI"),
    ("extrinsics.png",       "Camera Extrinsics"),
    ("ext_frustrum.png",     "Camera Frustums"),
    ("bundle.png",           "Bundle Adjustment"),
    ("3d_plot_bundle.png",   "3D Camera Trajectory"),
]

cols = 2
rows = (len(pipeline) + cols - 1) // cols


fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()

for ax, (fname, title) in zip(axes, pipeline):
    img_path = os.path.join(IMG_DIR, fname)
    img = Image.open(img_path)
    ax.imshow(img)
    ax.axis("off")

for ax in axes[len(pipeline):]:
    ax.axis("off")

plt.suptitle("Pipeline Overview (Laptop)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
